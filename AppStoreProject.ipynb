```python
#changing to appropriate datatypes
df['App_Name'] = df['App_Name'].astype('str')
df['Primary_Genre'] = df['Primary_Genre'].astype('category')
df['Content_Rating'] = df['Content_Rating'].astype('category')
df['Released']= pd.to_datetime(df['Released'])
df['Updated']=pd.to_datetime(df['Updated'])
df['Free'] = df['Free'].astype('category')
df['DeveloperId'] = df['DeveloperId'].astype('str')
```

```python
#function to visualize distribution and outliers
def uni_num(df,col):
    fig, (ax1,ax2,) = plt.subplots(ncols=2,figsize=(10,5))
    sns.kdeplot(data= df, x=col,ax=ax1,color='red')
    ax1.set_title("KDE Plot of " + col)
    sns.boxplot(df[col],orient='horizontal',ax=ax2,color='orange')
    ax2.set_title("Boxplot of " + col)
    plt.tight_layout()
    plt.show()
```

```python
def uni_num(df,col):
    #getting the stats into dictionary
    dctry = df[col].describe().to_dict()
    #Central Tendency and Spread summary
    maxi = round(dctry['max'],2)
    mini = round(dctry['min'],2)
    mean = round(dctry['mean'],2)
    stdd = round(dctry['std'],2)
    ran = round(dctry['max']-dctry['min'],2)
    md=df[col].mode()[0]
    sk = round(df[col].skew(),2)
    #Quantile
    q1 = round(dctry['25%'],2)
    q2 = round(dctry['50%'],2)
    q3 = round(dctry['75%'],2)
    #Size of plot
    fig, (ax1,ax2,) = plt.subplots(ncols=2,figsize=(15,5))
    fig.suptitle(col,fontsize=15)
    #KDE Plot
    sns.kdeplot(data= df, x=col,ax=ax1,fill=True,color='red')
    ax1.set_title('Central Tendency and Spread:\n' + f'Mean={mean} Mode={md} Standard Deviation ={stdd} Skewness={sk}\nMax={maxi} Min={mini} Range={ran}')
    ax1.set_xlabel("Kdeplot")
    ax1.set_ylabel('') 
    ax1.set_yticks([])
    #Outliers
    sns.boxplot(df[col],orient='horizontal',ax=ax2,color='orange')
    ax2.set_title('Quantile:\n' + f'First Quantile={q1} Second Quantile(Median)={q2}\n Third Quantile={q3} ')
    ax2.set_xlabel("Boxplot")
    ax2.set_ylabel('')
    ax2.set_yticks([])
    plt.tight_layout()
    plt.show()
```

```python
#Converting the app size from bytes to MB
df['Size_MB'] = df['Size_Bytes'] / 1024 / 1024
df['Size_MB'] = df['Size_MB'].round(2)
```

```python
#Creating bucket of sizes of apps
bins=[0,200,400,600,800,1000]
labels=['0-200','200-400','400-600','600-800','>=1000']
df['Bucket']=pd.cut(df['Size_MB'],bins,labels=labels)
```

```python
#Checking average user rating by bucket of app size
size_v_rating=pd.pivot_table(df,index='Bucket',values='Average_User_Rating',aggfunc='mean')
size_v_rating.plot(kind='line',title='Average Rating by Size(MB)',legend=False,xlabel='Groups of app sizes')
plt.tight_layout()
```

**Observation:**
* Above plot show that apps with sizes between 0 and 200 MB have an average user rating above 1.5.
* Apps with sizes greater than 200 MB tend to have an average user rating around 2.5. As large app size means they are more features and functionality satisfying different user needs.
